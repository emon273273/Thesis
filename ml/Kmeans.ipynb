```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/kmeans/student_clustering.csv")
df
```

```python
df.shape #cgpa and iq
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

#using elbow method

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
wcss
```

```python
#plot
plt.plot(range(1,11),wcss)
```

```python
X=df.iloc[:,:].values
km=KMeans(n_clusters=4)
y_means=km.fit_predict(X)

y_means
```

```python
df['cluster']=y_means
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
plt.scatter(df1.cgpa,df1['iq'],color='green')
plt.scatter(df2.cgpa,df2['iq'],color='red')
plt.scatter(df3.cgpa,df3['iq'],color='black')
plt.scatter(df4.cgpa,df4['iq'],color='blue')
plt.xlabel('Cgpa')
plt.ylabel("Iq Label")
plt.show()
```
